# Exploratory data analysis in Python - EDA Process

---

## Introduction

**What is Exploratory Data Analysis ?**

- Exploratory Data Analysis (EDA) is the process of analyzing and visualizing datasets to uncover patterns, trends, and relationships within the data before applying statistical modeling. Its primary goals include understanding the data structure, identifying anomalies and missing values, and generating hypotheses for further analysis. Common techniques used in EDA include descriptive statistics, data visualization (such as histograms and scatter plots), and correlation analysis. EDA is essential for gaining insights and guiding data-driven decisions in any analysis project.

**How to perform Exploratory Data Analysis ?**

- To perform Exploratory Data Analysis (EDA), start by importing necessary libraries (like Pandas and Matplotlib) and loading your dataset. Understand the data structure using methods like .head(), .info(), and .describe(). Check for missing values with .isnull().sum(), and clean the data by addressing duplicates or inconsistencies. Use visualizations, such as histograms, scatter plots, and box plots, to explore distributions and relationships between variables. Finally, summarize your findings to generate insights that can guide further analysis.

## 12. Outlier Detection
- For numerical columns, plot distributions using histograms, box plots, or sns.pairplot() to detect potential outliers visually. Calculate the Interquartile Range (IQR) or Z-scores for statistical outlier detection.


```python
numerical_columns = df_non_ev.select_dtypes(include=['int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df[col], kde=True, bins=40)
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

#### <span style="color:#ADD8E6">Notes:</span>
<span style="color:#ADD8E6">
 
I use IQR method.
</span>


```python
outliers = {}

for col in numerical_columns:
    Q1 = df_non_ev[col].quantile(0.25)
    Q3 = df_non_ev[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Find outliers
    outliers[col] = df_non_ev[(df_non_ev[col] < lower_bound) | (df_non_ev[col] > upper_bound)]
    
# Display the number of outliers for each numerical column
outlier_counts = {col: len(outliers[col]) for col in outliers}
 
outlier_counts 
 
```

## 13. Outlier Handling 
- Address outliers using methods like capping (clipping to upper/lower bounds), removing, or binning values. Consider keeping outliers if they represent meaningful data. Otherwise dropping the missing or null values.

---

#### <span style="color:#ADD8E6">Notes:</span>
<span style="color:#ADD8E6">

I chose to use Capping Method, most of the values seem correct, I think using capping will be useful for this dataset.
There are outliers but they represent valid data points (like high-performance cars), the capping method is still valid, as it keeps those extreme values but limits their impact on the overall analysis. 
</span>

```python
df_capped = df_non_ev.copy()
for col in numerical_columns:
    lower_percentile = df_non_ev[col].quantile(0.01)
    upper_percentile = df_non_ev[col].quantile(0.99)
    # Cap values at the 1st and 99th percentiles
    df_capped[col] = df_non_ev[col].clip(lower=lower_percentile, upper=upper_percentile)
```

## 14. Exploratory Visualization:
- Plot histograms for distribution, bar plots for categorical variables, and pair plots to observe relationships between features. Use sns.heatmap() for visual correlation analysis between numerical columns.

```python
# Before capping
numerical_columns = df_non_ev.select_dtypes(include=['int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df_non_ev[col], kde=True, bins=40)
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()

# After capping
numerical_columns = df_capped.select_dtypes(include=['int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 4, i)
    sns.histplot(df_capped[col], kde=True, bins=40)
    plt.title(f'Histogram of {col}')

plt.tight_layout()
plt.show()
```

```python
categorical_columns = df_capped.select_dtypes(include=['category']).columns
 
# Loop through each categorical column and plot the countplot individually
for col in categorical_columns:
    plt.figure(figsize=(10, 6))  # Set the figure size
    sns.countplot(x=df_capped[col], palette='Set2')  # Create the countplot
    plt.title(f"Bar Plot of {col}")  # Set the title
    plt.xticks(rotation=45)  # Rotate x-axis labels for readability
    plt.show()  # Show the plot
```

```python
pair_columns = ['Engine Horsepower', 'Highway Miles Per Gallon', 'City Miles Per Gallon', 'Manufacturer\'s Suggested Retail Price']

# Plot pair plots to observe relationships between selected numerical columns
sns.pairplot(df_capped[pair_columns], kind='scatter', plot_kws={'alpha': 0.5})
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df_capped[numerical_columns].corr()

# Plot heatmap to observe correlations between numerical features
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, square=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
